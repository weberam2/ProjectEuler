```python
import pandas
from collections import Counter
```

```python
encr1="BT JPX RMLX PCUV AMLX ICVJP IBTWXVR CI M LMT'R PMTN, MTN" 
encr2="YVCJX CDXV MWMBTRJ JPX AMTNGXRJBAH UQCT JPX QGMRJXV CI JPX"
encr3="YMGG CI JPX HBTW'R QMGMAX; MTN JPX HBTW RMY JPX QMVJ CI JPX"
encr4="PMTN JPMJ YVCJX. JPXT JPX HBTW'R ACUTJXTMTAX YMR APMTWXN"
encrypted=' '.join([encr1,encr2,encr3, encr4])
print(encrypted)
```

```python
encryptednospace = encrypted.replace(" ", "")
letter_counts = Counter(encryptednospace)
df = pandas.DataFrame.from_dict(letter_counts, orient='index')
df.plot(kind='bar')
```
